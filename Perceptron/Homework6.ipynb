# Perceptron Model

```python
# Perceptron Model

import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt

input  = [ [-1,0],[-1,-1],[1,1],[1,0] ]
output =  [  0    , 0     , 1   , 1    ]

for i in range( len(input) ):
    input[ i ].insert(0,1)

x = np.array(input)
w = np.array( [0,0,0] )
alpha = 0.2
n = len(input)

for i in range( n ):
    h = w.dot( input[i] )

    if h >= 0:
        h = 1
    else: 
        h = 0

    w = w + alpha*( output[i] - h ) * x[i]

    print(w)

print(x.dot(w))
```

```python
# Question 2
alpha = 0.25
def perceptron(input_x,input_y):

    x = np.array( input_x )
    w = np.array( [0.1,0.1,0.1] )

    for j in range(5):

        for i in range( len(input_x) ):
            h = w.dot( input_x[i] )

            if h >= 0:
                h = 1
            else: 
                h = 0

            w = w + alpha*(0.5**j)*( input_y[i] - h ) * x[i]

    plt.scatter( [ x[i][1] for i in range(n) ] ,  [ x[i][2] for i in range(n) ] )

    x_ = np.linspace(-0.1,1.1,100)
    y_ = np.linspace(-0.1,1.1,100)
    x_,y_ = np.meshgrid(x_,y_)
    z = w[0] + w[1]*x_ + w[2]*y_
    plt.contour(x_,y_,z,0)
    plt.show() 

    return w

input = [ [0,0], [0,1], [1,0], [1,1] ]
for i in range( len(input) ):
    input[ i ].insert(0,1)
    
out_and = [0,0,0,1]
out_or  = [0,1,1,1]
out_xor = [0,1,1,0]

perceptron(input,out_and)
perceptron(input,out_or)
perceptron(input,out_xor)
```
